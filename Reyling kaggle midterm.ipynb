```python
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model, datasets
import pandas as pd
%pylab inline
```

```python
train = pd.read_csv("midterm_train.csv")
```

```python
train
```

```python
train.describe()
```

```python
import re
def clean_loan(x):
        x = x.strip('$')
        return int(x.replace(',',''))
    
train['Loan_Amount_Requested']=train['Loan_Amount_Requested'].map(clean_loan)
```

```python
train['Loan_Amount_Funded'] = train['Loan_Amount_Funded'].map(clean_loan)
```

```python
train['Loan_Amount_Funded_by_Investors'] = train['Loan_Amount_Funded_by_Investors'].map(clean_loan)
```

```python
train.dtypes
```

```python
def clean_interest(x):
    temp = str(x).strip('%')
    return float(temp)/100
    
train['Interest_Rate']=train['Interest_Rate'].map(clean_interest)
```

```python
train
```

```python
train.describe()
```

```python
mean_Int_rate = mean(train['Interest_Rate'])
mean_Int_rate
train.Interest_Rate = train.Interest_Rate.fillna(value=mean_Int_rate)
```

```python
train.describe()
```

```python
y_train = train.pop('Interest_Rate')
y_train
```

```python
def describe_categorical(X):
    from IPython.display import display, HTML
    display(HTML(X[X.columns[X.dtypes == 'object']].describe().to_html()))
```

```python
describe_categorical(train)
```

```python
train.drop(['Loan_ID'], axis=1, inplace=True)
train.drop(['Borrower_ID'], axis=1, inplace=True)
```

```python
X_train=train
X_train.drop(['Month_Issued'], axis=1, inplace=True)
X_train.drop(['Current_Status'], axis=1, inplace=True)
X_train.drop(['Loan_Description'], axis=1, inplace=True)
X_train.drop(['Loan_Grade'], axis=1, inplace=True)
X_train.drop(['Employer'], axis=1, inplace=True)
```

```python
X_train.drop(['Purpose_Of_Loan'], axis=1, inplace=True)
X_train.drop(['Loan_Title'], axis=1, inplace=True)
X_train.drop(['Zip_3'], axis=1, inplace=True)
X_train.drop(['State'], axis=1, inplace=True)
X_train.drop(['Debt_To_Income'], axis=1, inplace=True)
```

```python
X_train.drop(['Number_Delinqueny_2yrs'], axis=1, inplace=True)
X_train.drop(['Month_Earliest_Credit'], axis=1, inplace=True)
X_train.drop(['Inquiries_Last_6Mo'], axis=1, inplace=True)
X_train.drop(['Months_Since_Deliquency'], axis=1, inplace=True)
X_train.drop(['Months_Since_Record'], axis=1, inplace=True)
X_train.drop(['Number_Open_Accounts'], axis=1, inplace=True)
```

```python
X_train.drop(['Public_Record_Count'], axis=1, inplace=True)
X_train.drop(['Revolving_Balance'], axis=1, inplace=True)
X_train.drop(['Revolving_Utilization'], axis=1, inplace=True)
X_train.drop(['Total_Accounts'], axis=1, inplace=True)
X_train.drop(['Initial_Status'], axis=1, inplace=True)
X_train.drop(['Outstanding_Principal'], axis=1, inplace=True)
```

```python
X_train.drop(['Outstanding_Principal_Remaining'], axis=1, inplace=True)
X_train.drop(['Total_Payment'], axis=1, inplace=True)
X_train.drop(['Total_Payment_Funded'], axis=1, inplace=True)
X_train.drop(['Total_Received'], axis=1, inplace=True)
X_train.drop(['Total_Received_Interest'], axis=1, inplace=True)
X_train.drop(['Total_Received_Late_Fees'], axis=1, inplace=True)
X_train.drop(['Recoveries'], axis=1, inplace=True)
X_train.drop(['Collection_Recovery_Fee'], axis=1, inplace=True)
X_train.drop(['Last_Payment_Received'], axis=1, inplace=True)
X_train.drop(['Last_Payment_Total'], axis=1, inplace=True)
X_train.drop(['Next_Payment_Date'], axis=1, inplace=True)
X_train.drop(['Last_Credit_Pulled'], axis=1, inplace=True)
X_train.drop(['Collections_12Mo_Exclude_Med'], axis=1, inplace=True)
X_train.drop(['Months_Since_Last_Derogatory'], axis=1, inplace=True)
X_train.drop(['Policy_Code'], axis=1, inplace=True)
```

```python
X_train.describe()
```

```python
mean_Income = mean(X_train['Annual_Income'])
mean_Income
X_train.Annual_Income = X_train.Annual_Income.fillna(value=mean_Income)
```

```python
X_train.describe()
```

```python
dummies = pd.get_dummies(X_train['Income_Verified'], prefix='Income_Verified')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Income_Verified'], axis=1, inplace=True)
X_train.drop(['Income_Verified_VERIFIED - income'], axis=1, inplace=True)
```

```python
dummies = pd.get_dummies(X_train['Initial_Loan_Term_Months'], prefix='Initial_Loan_Term_Months')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Initial_Loan_Term_Months'], axis=1, inplace=True)
X_train.drop(['Initial_Loan_Term_Months_ 36 months'], axis=1, inplace=True)
```

```python
dummies = pd.get_dummies(X_train['Sub_Loan_Grade'], prefix='Sub_Loan_Grade')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Sub_Loan_Grade'], axis=1, inplace=True)
X_train.drop(['Sub_Loan_Grade_0'], axis=1, inplace=True)
```

```python
dummies = pd.get_dummies(X_train['Length_Employed'], prefix='Length_Employed')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Length_Employed'], axis=1, inplace=True)
X_train.drop(['Length_Employed_1 year'], axis=1, inplace=True)
```

```python
dummies = pd.get_dummies(X_train['Home_Owner?'], prefix='Home_Owner')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Home_Owner?'], axis=1, inplace=True)
X_train.drop(['Home_Owner_0'], axis=1, inplace=True)
```

```python
dummies = pd.get_dummies(X_train['Payment_Plan_Ind'], prefix='Payment_Plan')
X_train = pd.concat([X_train, dummies], axis=1)
X_train.drop(['Payment_Plan_Ind'], axis=1, inplace=True)
X_train.drop(['Payment_Plan_n'], axis=1, inplace=True)
```

```python
def printall(X, max_rows=10):
    from IPython.display import display, HTML
    display(HTML(X.to_html(max_rows=max_rows)))
    
printall(X_train)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_train.shape
```

```python
linearRegression = linear_model.LinearRegression()
```

```python
linearRegression.fit(X_train, y_train)
```

```python
print("Variance: %f" %linearRegression.score(X_train, y_train))
```
