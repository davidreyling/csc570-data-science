```python
import pandas as pd
```

```python
df = pd.read_csv("breast_cancer.csv")
df
```

```python
df.describe()
```

```python
df = df.drop(['Unnamed: 0', 'id number'], axis=1)
df
```

```python
y = df.pop("malignant")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn import cross_validation
import math

%pylab inline
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=.2, random_state=42)
```

```python
n_estimators = [300,400,500]
max_features = ['auto', 'sqrt', 'log2']
min_samples_split = [3,5,7]

rfc = RandomForestClassifier(n_jobs=1)
estimator = GridSearchCV(rfc, 
                            dict(n_estimators=n_estimators, 
                                max_features=max_features, 
                                min_samples_split=min_samples_split), 
                                cv=None, n_jobs=-1)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
best_rfc = estimator.best_estimator_
```

Accuracy

```python
accuracy = accuracy_score(y_test, best_rfc.predict(X_test))
print "Accuracy: ", accuracy
```

Precision & Recall

```python
print classification_report(y_test, best_rfc.predict(X_test))
```

AUC

```python
roc = roc_auc_score(y_test, best_rfc.predict_proba(X_test)[:,1])
print "AUC Score: ", roc
```

```python
scores = cross_validation.cross_val_score
scores = cross_validation.cross_val_score(best_rfc, df, y, cv=10)
scores
```

```python
mean_score = scores.mean()
std_dev = scores.std()
std_error = scores.std() / math.sqrt(scores.shape[0])
ci = 2.262 * std_error
lower_bound = mean_score - ci
upper_bound = mean_score + ci

print "Score is %f +/- %f" %(mean_score, ci)
print "95 percent probability that if this experiment were repeated over and over the average score would be between %f and %f" %(lower_bound, upper_bound)
```

The K-fold CV score and the single holdout AUC were fairly close.  The accuracy of the model was  96.4% accurate, and our K-Fold score was between 95% accurate to 99% accurate.  On the whole K-Fold would typically give a slightly better score averaging around .2 percentage point higher.

Our model will correctly identify with a person has breast cancer is 95%.  The model will identify a person as having breast cancer when in fact they do not have breast cancer 5% of the time.  For our model of the people it identifies as having breast cancer 93% of those people will actually have breast cancer.

The AUC for our model is .996.  This is a measure of how good our model is.  .996 is near perfect so our model is pretty good.  This will help us determine a threshold for a sample size where we can actually get a good percentage of our predictions correct.
